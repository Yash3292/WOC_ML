```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from logistic_regression import Logistic_Regression
```

```python
df = pd.read_csv('binary_classification_train.csv',header = None)
```

```python
df
```

```python
X = np.array(df.iloc[1:,1:21], dtype=np.float64)
y = np.array(df.iloc[1:,21], dtype=np.float64)
```

```python
model = Logistic_Regression()
```

```python
X_train,X_test,y_train,y_test = model.test_train_split(X, y, 0.4)
```

```python
model.train()
```

```python
model.fit(X_train, y_train)
```

```python
model.plot_cost()
```

```python
yhat_train = model.predict(X_train)
print(model.accuracy(yhat_train, y_train))
yhat_test = model.predict(X_test)
print(model.accuracy(yhat_test, y_test))
```

```python
print(model.f1_score(yhat_train, y_train))
print(model.f1_score(yhat_test, y_test))
```
