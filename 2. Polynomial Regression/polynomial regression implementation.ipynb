```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from linear_regression import Linear_Regression
from polynomial import CreatePolynomialFeatures
```

```python
df = pd.read_csv('polynomial_regression_train.csv',header = None)
```

```python
X = np.array(df.iloc[1:,1:6], dtype=np.float64)
y = np.array(df.iloc[1:,6], dtype=np.float64)
```

```python
X = CreatePolynomialFeatures().poly_transform(X, 6)
```

```python
model = Linear_Regression()
```

```python
X_train,X_test,y_train,y_test = model.test_train_split(X, y, 0.4)
```

```python
model.train()
```

```python
model.fit(X_train, y_train)
```

```python
model.plot_cost()
```

```python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
yhat = model.predict(X_train)
J_train = model.mse(yhat, y_train)
print("Train cost:", J_train)
yhat = model.predict(X_test)
J_test = model.mse(yhat, y_test)
print("Test cost:", J_test)
```
